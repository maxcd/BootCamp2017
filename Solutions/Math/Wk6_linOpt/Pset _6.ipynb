### Problem 1.
max $5x - y $ s.t. c1, c2, c3

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as opt
```

```python
c1 = lambda x: 2/3 * x + 4 / 3
c2 = lambda x: 1/6*(x-1)
c3 = lambda x: 6-x

x = np.linspace(0, 6, 12)

C1 = c1(x)
C2 = c2(x)
C3 = c3(x)

plt.plot(x, c1(x), ':r', label='c1',)
plt.plot(x, c2(x), ':g',label = 'c2')
plt.plot(x, c3(x), ':k',label = 'c3')
plt.plot(x, np.zeros(12), 'b')
plt.plot(np.zeros(12), x, 'b')
plt.plot(0, 6, 'k+', markersize=12, label="maximizer")
plt.text(0.5, 3.25, 'feasible set' )
plt.title('optimal value of 24 at (0,6)')
plt.legend()
plt.show()
```

## Problem 8.17


Let the primal problem be given by 
$$max \, c^Tx$$
$$ Ax \leq b$$
$$  x \geq 0$$

with the corresponding dual being 
$$max \, b^Ty$$
$$ A^Ty \geq c$$
$$  y \geq 0$$

taking the dial of the dual yields
$$max \, c^Tz$$
$$ Az \leq b$$
$$  z \geq 0$$
which looks is the same as the primal problem and hence $z=x$.
